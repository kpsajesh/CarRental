```python
#from car_rental import CarRental
from IPython.display import clear_output 
from datetime import datetime

def main():    
    objCarRental = CarRental(250)
    objCustomer = Customer()
    
    maxStock = 0
    stockAvailable = 0
    maxCarsCanReturn = 0
    billAmount = 0

    DailyRentedDays = 0
    
    while True:        
        print(f'\n0. To show available cars.')
        print(f'1. To request car on hourly basis. Tariff Rs{objCarRental.hourlyTariff}/hour.')
        print(f'2. To return car on hourly basis.')
        print(f'3. To request car on daily basis. Tariff Rs{objCarRental.dailyTariff}/day.')
        print(f'4. To return car on daily basis.')
        print(f'5. To request car on weekly basis.Tariff Rs{objCarRental.weeklyTariff}/week.')
        print(f'7. Exit.')
    
        userMainInput = UserInputNumberBetween(0,7,'\nEnter your choice:')
        
        if userMainInput == 0:
            clear_output()
            stockAvailable = objCarRental.ShowAvailableStock()
            print(f'Currently available cars to rent: {stockAvailable}')
        elif userMainInput == 1:
            print('You selected the HOURLY RENTAL HIRE option.')
            stockAvailable = objCarRental.ShowAvailableStock()
            if stockAvailable > 0:
                objCustomer.RequestCarHourly(stockAvailable)
                print(f'(objCustomer.hourlyCarRequestCount {objCustomer.hourlyCarRequestCount}')
                if objCarRental.RentCar(objCustomer.hourlyCarRequestCount) == True:
                    clear_output()
                    print(f'Hourly Hire Success: {objCustomer.hourlyCarRequestCount} car/s issued for hourly rent from {objCustomer.hourlyFromHour}th hour at daily tariff of Rs{objCarRental.hourlyTariff}/hour')
                    stockAvailable = objCarRental.ShowAvailableStock()
                    print(f'Currently available cars to rent: {stockAvailable}.')
            else:
                print(f'Sorry, Our all cars on the road. There are no cars available for rent.')
            
        elif userMainInput == 2:
            print('You selected the HOURLY RENTAL HIRE option.')
            maxStock = objCarRental.ShowMaxStock()
            stockAvailable = objCarRental.ShowAvailableStock()
            maxCarsCanReturn = maxStock - stockAvailable ##### Remove 5 count   
            if maxCarsCanReturn > 0:
                HourlyRentedHours = objCustomer.ReturnCarHourly(maxCarsCanReturn)
                if HourlyRentedHours > 0:
                    clear_output()
                    billAmount = objCarRental.ReturnHourly(HourlyRentedHours, objCustomer.hourlyCarReturnCount)
                    print('Thanks you for using our service.')
                    print(f'Your bill amount is Rs{billAmount}.')
                    stockAvailable = objCarRental.ShowAvailableStock()
                    print(f'Currently available cars to rent: {stockAvailable}')
            else:
                print(f'Sorry, No cars rented yet, so no cars can return.')
                
        elif userMainInput == 3:
            print('You selected the DAILY RENTAL HIRE option.')
            stockAvailable = objCarRental.ShowAvailableStock()
            if stockAvailable > 0:
                objCustomer.RequestCarDaily(stockAvailable)
                if objCarRental.RentCar(objCustomer.dailyCarRequestCount) == True:
                    clear_output()
                    print(f'Daily Hire Success: {objCustomer.dailyCarRequestCount} car/s issued for daily rent from {objCustomer.dailyFromDate} at daily tariff of Rs{objCarRental.dailyTariff}/day')
                    stockAvailable = objCarRental.ShowAvailableStock()
                    print(f'Currently available cars to rent: {stockAvailable}.')
            else:
                print(f'Sorry, Our all cars on the road. There are no cars available for rent.')
            
        elif userMainInput == 4:
            print('You selected the DAILY RENTAL RETURN option.')
            maxStock = objCarRental.ShowMaxStock()
            stockAvailable = objCarRental.ShowAvailableStock()
            maxCarsCanReturn = maxStock - stockAvailable ##### Remove 5 count            
            if maxCarsCanReturn > 0:
                DailyRentedDays = objCustomer.ReturnCarDaily(maxCarsCanReturn)
                if DailyRentedDays > 0:
                    clear_output()
                    billAmount = objCarRental.ReturnDaily(DailyRentedDays, objCustomer.dailyCarReturnCount)
                    print('Thanks you for using our service.')
                    print(f'Your bill amount is Rs{billAmount}.')
                    stockAvailable = objCarRental.ShowAvailableStock()
                    print(f'Currently available cars to rent: {stockAvailable}')
            else:
                print(f'Sorry, No cars rented yet, so no cars can return.')
        elif userMainInput == 5:
            print('5 Todo')
        elif userMainInput == 6:
            print('6 Todo')
        elif userMainInput == 7:
            print('HAPPY SERVING YOU!!! BYE.')
            break
        else:
            print('Invalid selection, please enter between 1 & 7.')
        

#Validation: Number validation for a range
def UserInputNumberBetween(FromNum,ToNum,strInputMsg):    
    #from IPython.display import clear_output    
    
    ToNumEnd = ToNum + 1
    strInputMsg = strInputMsg +': '
    while True:
        a = input(strInputMsg)
        if a in map(str,range(FromNum,ToNumEnd)):
            a = int(a)
            break
        else:
            print('Wrong input, please enter a number between', FromNum,'and',ToNum,'.')
    return a


    
if __name__ == "__main__":
    main()
        
```

```python
from datetime import datetime, timedelta
from IPython.display import clear_output 

class CarRental:
    def __init__(self, stock=0): #Constructor car Rental
        self.stock = stock
        self.runningStock = stock
        self.hourlyTariff = 200 #Rs200/hour
        self.dailyTariff = 1500 #Rs1500/hour
        self.weeklyTariff = 7500 #Rs7500/hour

        self.billAmount = 0

    def ShowAvailableStock(self): #Display the running stock available for rent 
        return self.runningStock
    
    def ShowMaxStock(self): #Display the max stock
        return self.stock

    def RentCar(self,rentCount): #Issue cars in hourly/daily/weekly rent basis
        self.runningStock -= rentCount
        return True
    def ReturnDaily(self,daysUsed,noofCars):
        self.billAmount = daysUsed * noofCars * self.dailyTariff
        self.runningStock += noofCars
        return self.billAmount

    def ReturnHourly(self,hoursUsed,noofCars):
        self.billAmount = hoursUsed * noofCars * self.hourlyTariff 
        self.runningStock += noofCars
        return self.billAmount


#*********************************************************************************************************   
class Customer():
    def __init__(self):        
        self.dailyCarRequestCount = 0 #Car request count - Daily
        self.dailyFromDate = 'NoDate' #Car request From date - Daily
        self.dailyCarReturnCount = 0  #Car return count - Daily
        self.dailyToDate = 'NoDate'   #Car return To date - Daily
        self.DailyHiredDays = 0       #No. of days hired

        self.hourlyCarRequestCount = 0 #Car request count - Hourly
        self.hourlyFromHour = 0        #Car request From hour - Hourly
        self.hourlyCarReturnCount = 0  #Car return count - Hourly
        self.hourlyToHour = 0          #Car return To hour - Hourly
        self.hourlyHiredhours = 0      #No. of hours hired
        
    def RequestCarHourly(self,carStock):   #Method to request cars on DAILY RENTAL basis     
        #Getting and validating the no. of car request for rent
        print(f'Currently available cars to rent:{carStock}')
        strCountCarMsg = 'HOURLY HIRE: How many cars you would like to rent? Request between 1 and '+ str(carStock) + '.'
        self.hourlyCarRequestCount = UserInputNumberBetween(1,carStock,strCountCarMsg)

        #Getting and validating the from hour of rent
        strCountCarMsg = 'HOURLY HIRE: Enter from hour of hire.? Request between 1 and 24.'
        self.hourlyFromHour = UserInputNumberBetween(1,24,strCountCarMsg) 
    
    def ReturnCarHourly(self,maxCarsCanReturn):
        #Getting and validating the no. of car return for rent         
        print(f'Total no. of cars can return: {maxCarsCanReturn}')
        strCountCarMsg = 'HOURLY HIRE RETURN: How many cars you would like to return? Request between 1 and '+ str(maxCarsCanReturn) + '.'
        self.hourlyCarReturnCount = UserInputNumberBetween(1,maxCarsCanReturn,strCountCarMsg)

        #Getting and validating the from hour of rent
        strCountCarMsg = 'HOURLY HIRE RETURN: Enter from hour of hire.? Request between 1 and 24.'
        self.hourlyFromHour = UserInputNumberBetween(1,24,strCountCarMsg) 

        #Getting and validating the from hour of rent
        strCountCarMsg = 'HOURLY HIRE RETURN: Enter end hour of hire.? Request between 1 and 24.'
        self.hourlyToHour = UserInputNumberBetween(1,24,strCountCarMsg) 

        HourlyHiredhours = self.hourlyToHour - self.hourlyFromHour
        if HourlyHiredhours > 0:
            return HourlyHiredhours
        else:
            print('HOURLY HIRE RETURN: Error. Recheck the from hour and end hour.')
            return 0
    
    def RequestCarDaily(self,carStock):   #Method to request cars on DAILY RENTAL basis     
        #Getting and validating the no. of car request for rent
        print(f'Currently available cars to rent:{carStock}')
        strCountCarMsg = 'DAILY HIRE: How many cars you would like to rent? Request between 1 and '+ str(carStock) + '.'
        self.dailyCarRequestCount = UserInputNumberBetween(1,carStock,strCountCarMsg)

        #Getting and validating the from date of rent
        while True:            
            date1 = input('DAILY HIRE: Enter from date of hire.')
            if DateValidation(date1) == True:
                self.dailyFromDate = date1
                break
            else:
                print('Daily Hire: Error. Invalid from date of hire')
       
    def ReturnCarDaily(self,maxCarsCanReturn):
        #Getting and validating the no. of car return for rent
        print(f'Total no. of cars can return: {maxCarsCanReturn}')
        strCountCarMsg = 'DAILY HIRE RETURN: How many cars you would like to retun? Return between 1 and '+ str(maxCarsCanReturn) + '.'
        self.dailyCarReturnCount = UserInputNumberBetween(1,maxCarsCanReturn,strCountCarMsg)

        #Getting and validating the from date of rent
        while True:            
            date1 = input('DAILY HIRE RETURN: Enter from date of hire.')
            if DateValidation(date1) == True:
                self.dailyFromDate = date1
                break
            else:
                print('DAILY HIRE RETURN: Error. Invalid from date of hire')
        
        #Getting and validating the to date of rent
        while True:            
            date2 = input('DAILY HIRE RETURN: Enter end date of hire.')
            if DateValidation(date2) == True:
                self.dailyToDate = date2
                break
            else:
                print('DAILY HIRE RETURN: Error. Invalid end date of hire.')

        #Getting the no. of days between the from and to dates
        DailyHiredDays = DaysBetween(date2,date1)
        if DailyHiredDays > 0:
            return DailyHiredDays
        else:
            print('DAILY HIRE RETURN: Error. Recheck the from date and end date.')
            return 0

# Validation: Number validation for a range
def UserInputNumberBetween(FromNum,ToNum,strInputMsg):  
    ToNumEnd = ToNum + 1
    strInputMsg = strInputMsg +': '
    while True:
        a = input(strInputMsg)
        if a in map(str,range(FromNum,ToNumEnd)):
            a = int(a)
            break
        else:
            print('Error, please enter a number between', FromNum,'and',ToNum,'.')
    return a
        
# Validation: Date
def DateValidation(date_text):
    try:
        if date_text != datetime.strptime(date_text, "%d/%m/%Y").strftime('%d/%m/%Y'):
            raise ValueError
        
        #weeklyFromDate = date_text
        return True
    except ValueError:
        return False

# Validation: Number of days between 2 dates
def DaysBetween(d1, d2):
    d1 = datetime.strptime(d1, "%d/%m/%Y")
    d2 = datetime.strptime(d2, "%d/%m/%Y")
    if d1 >= d2:
        return abs((d2 - d1).days) + 1
    else:
        return -1
```
